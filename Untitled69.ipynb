```python
!pip install wbgapi
import wbgapi as wb
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
country_codes = ['GTM','UMC','WLD'] # GTM:Guatemala, UMC: upper middle income ,WLD: world
indicators = {
   'NY.GDP.PCAP.KD': 'GDP_Per_Capita_Const',     # Standard of Living
    'SL.TLF.CACT.ZS': 'Labor_Part_Rate',          # Labor Supply
    'SL.UEM.TOTL.ZS': 'Unemployment_Rate',        # Labor Friction
    'SL.TLF.TOTL.IN': 'Labor_Force_Total',        # Total Workers
    'NY.GDP.MKTP.KD': 'GDP_Const',                # Economic Size
    'FP.CPI.TOTL.ZG': 'Inflation_CPI',            # Stability
    'NY.GDS.TOTL.ZS': 'Gross_Dom_Savings',        # Savings
    'NE.GDI.TOTL.ZS': 'Gross_Cap_Formation',      # Investment
    'NE.EXP.GNFS.ZS': 'Exports_GDP',              # Exports
    'NE.IMP.GNFS.ZS': 'Imports_GDP',              # Imports
    'GC.TAX.TOTL.GD.ZS': 'Tax_Rev_GDP',           # Gov Revenue
    'NE.CON.GOVT.ZS': 'Gov_Exp_GDP'               # Gov Spending
}
```

```python
print("Fetching data...")
df_raw = wb.data.DataFrame(indicators,economy=country_codes,time=range(2000,2025))
df=df_raw.T
df.index = df.index.str.replace('YR','').astype(int)
df.index.name = 'Year'
df = df.sort_index()
df.rename(columns=indicators,level=1,inplace=True)
df_gtm = df.xs('GTM',axis=1,level=0).copy()
```

```python
df_gtm['Natural_Rate_Unemp']=df_gtm['Unemployment_Rate'].rolling(window=5).mean()
df_gtm['Productivity']=df_gtm['GDP_Const']/df_gtm['Labor_Force_Total']
df_gtm['NCO']=df_gtm['Exports_GDP']-df_gtm['Imports_GDP']
df_gtm['Budget_Balabce']=df_gtm['Tax_Rev_GDP']-df_gtm['Gov_Exp_GDP']
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index,df[('GTM','GDP_Per_Capita_Const')],label='Guatemala',color='blue',linewidth=3)
plt.title('Benchmarking Development: Real GDP Per Capita(2000-2023)',fontsize=16)
plt.ylabel('Constant 2015 US$')
plt.xlabel('Year')
plt.legend()
plt.grid(True,alpha=0.3)
plt.show()
```

This plot show GDP per capita for Guatemala only. From this plot, we will see a massive growth because the lin goes up significantly.

```python
plt.figure(figsize=(12,6))
plt.plot(df.index,df[('GTM','GDP_Per_Capita_Const')],label='Guatemala',color='blue',linewidth=3)
plt.plot(df.index,df[('UMC','GDP_Per_Capita_Const')],label='Upper Middle Income(Avg)',color='red',linestyle='--',linewidth=2)
plt.plot(df.index,df[('WLD','GDP_Per_Capita_Const')],label='World Average',color='black',linestyle=':',linewidth=2)
plt.title('Benchmarking Development: Real GDP Per Capita(2000-2023)',fontsize=16)
plt.legend()
plt.grid(True,alpha=0.3)
plt.show()
```

After adding world and upper middel income's lines, the increase of Guatemala become float.

The following plot is created by AI and this is the prompt:                    
"I have a DataFrame 'df_gtm' containing economic indicators for Guatemala.
I need to create a **2x3 Executive Dashboard** summarizing the economy using Matplotlib/Seaborn.

**The Variables:**
* Real GDP: 'GDP_Const'
* Inflation: 'Inflation_CPI'
* Unemployment: 'Unemployment_Rate'
* Tax Revenue: 'Tax_Rev_GDP'
* Gov Expenditure: 'Gov_Exp_GDP'
* Exports: 'Exports_GDP'
* Imports: 'Imports_GDP'
* Savings: 'Gross_Dom_Savings'
* Investment: 'Gross_Cap_Formation'

**Requirements:**
1. Top Left: Real GDP (Line chart).
2. Top Middle: Inflation Rate (Bar chart with a horizontal line at 0).
3. Top Right: Unemployment Rate (Line chart).
4. Bottom Left: Fiscal Balance (Fill area between Tax Revenue and Gov Expenditure).
5. Bottom Middle: Trade Balance (Fill area between Exports and Imports).
6. Bottom Right: Savings vs Investment (Dual lines).

**Style:** * Use plt.style.use('dark_background').
* Add a main title: 'Guatemala Economic Snapshot'.
* Use tight_layout() to prevent overlap."

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Style
plt.style.use('default')
sns.set_context('talk')

# Create 2x3 dashboard
fig, axes = plt.subplots(2, 3, figsize=(18, 10), sharex=True)
fig.suptitle('Guatemala Economic Snapshot', fontsize=22, y=1.02)

# 1. Top Left: Real GDP
axes[0, 0].plot(df_gtm.index, df_gtm['GDP_Const'], linewidth=2.5)
axes[0, 0].set_title('Real GDP')
axes[0, 0].set_ylabel('Constant GDP')
axes[0, 0].grid(True, alpha=0.2)

# 2. Top Middle: Inflation Rate
axes[0, 1].bar(df_gtm.index, df_gtm['Inflation_CPI'], alpha=0.8)
axes[0, 1].axhline(0, linestyle='--', linewidth=1.5)
axes[0, 1].set_title('Inflation Rate')
axes[0, 1].set_ylabel('%')
axes[0, 1].grid(True, axis='y', alpha=0.2)

# 3. Top Right: Unemployment Rate
axes[0, 2].plot(df_gtm.index, df_gtm['Unemployment_Rate'], linewidth=2.5)
axes[0, 2].set_title('Unemployment Rate')
axes[0, 2].set_ylabel('%')
axes[0, 2].grid(True, alpha=0.2)

# 4. Bottom Left: Fiscal Balance
tax = df_gtm['Tax_Rev_GDP']
gov = df_gtm['Gov_Exp_GDP']

axes[1, 0].plot(df_gtm.index, tax, label='Tax Revenue', linewidth=2)
axes[1, 0].plot(df_gtm.index, gov, label='Gov Expenditure', linewidth=2)
axes[1, 0].fill_between(
    df_gtm.index, tax, gov,
    where=(tax >= gov),
    alpha=0.3,
    interpolate=True,
    label='Fiscal Surplus'
)
axes[1, 0].fill_between(
    df_gtm.index, tax, gov,
    where=(tax < gov),
    alpha=0.3,
    interpolate=True,
    label='Fiscal Deficit'
)
axes[1, 0].set_title('Fiscal Balance')
axes[1, 0].set_ylabel('% of GDP')
axes[1, 0].legend(fontsize=9)
axes[1, 0].grid(True, alpha=0.2)

# 5. Bottom Middle: Trade Balance
exports = df_gtm['Exports_GDP']
imports = df_gtm['Imports_GDP']

axes[1, 1].plot(df_gtm.index, exports, label='Exports', linewidth=2)
axes[1, 1].plot(df_gtm.index, imports, label='Imports', linewidth=2)
axes[1, 1].fill_between(
    df_gtm.index, exports, imports,
    where=(exports >= imports),
    alpha=0.3,
    interpolate=True,
    label='Trade Surplus'
)
axes[1, 1].fill_between(
    df_gtm.index, exports, imports,
    where=(exports < imports),
    alpha=0.3,
    interpolate=True,
    label='Trade Deficit'
)
axes[1, 1].set_title('Trade Balance')
axes[1, 1].set_ylabel('% of GDP')
axes[1, 1].legend(fontsize=9)
axes[1, 1].grid(True, alpha=0.2)

# 6. Bottom Right: Savings vs Investment
axes[1, 2].plot(df_gtm.index, df_gtm['Gross_Dom_Savings'], linewidth=2.5, label='Savings')
axes[1, 2].plot(df_gtm.index, df_gtm['Gross_Cap_Formation'], linewidth=2.5, label='Investment')
axes[1, 2].set_title('Savings vs Investment')
axes[1, 2].set_ylabel('% of GDP')
axes[1, 2].legend(fontsize=9)
axes[1, 2].grid(True, alpha=0.2)

# X labels for bottom row
for ax in axes[1, :]:
    ax.set_xlabel('Year')

plt.tight_layout()
plt.show()
```
