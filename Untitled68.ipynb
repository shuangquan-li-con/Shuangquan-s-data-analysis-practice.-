```python
def get_real_wage(nominal_col,cpi_col):
  """
  Takes a nominal column and a CPI column.
  Returns a new column adjusted for inflation(Real Dollars).
  """
  cpi_today = cpi_col.iloc[-1]
  real_col = (nominal_col/cpi_col)*cpi_today
  return real_col
df['Real_Wage']=get_real_wage(df['Nominal_Wage'],df['CPI'])
print(df.tail())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Fetch ECIWAG using the existing fred object
eci = fred.get_series("ECIWAG")

# Ensure datetime index
wages.index = pd.to_datetime(wages.index)
eci.index = pd.to_datetime(eci.index)

# 2. Filter both series from 2015-01-01 onward
start_date = "2015-01-01"
wages_filtered = wages.loc[start_date:].copy()
eci_filtered = eci.loc[start_date:].copy()

# Drop missing values just in case
wages_filtered = wages_filtered.dropna()
eci_filtered = eci_filtered.dropna()

# 3. Rebase both to 100
wages_rebased = (wages_filtered / wages_filtered.iloc[0]) * 100
eci_rebased = (eci_filtered / eci_filtered.iloc[0]) * 100

# 4. Plot both series
plt.figure(figsize=(14, 7))

plt.plot(
    wages_rebased.index,
    wages_rebased,
    color="red",
    linestyle="--",
    linewidth=2,
    label="Standard Average - Biased"
)

plt.plot(
    eci_rebased.index,
    eci_rebased,
    color="green",
    linestyle="-",
    linewidth=3,
    label="ECI - Fixed Composition"
)

# 5. Annotate the 2020 divergence
wages_2020 = wages_rebased.loc["2020"]
eci_2020 = eci_rebased.loc["2020"]

wages_spike_date = wages_2020.idxmax()
wages_spike_value = wages_2020.max()

# Use .asof() to get the nearest valid quarterly ECI value for the monthly wages_spike_date
eci_same_date_value = eci_rebased.asof(wages_spike_date)

plt.annotate(
    "Artificial Spike",
    xy=(wages_spike_date, wages_spike_value),
    xytext=(pd.Timestamp("2018-06-01"), wages_spike_value + 6),
    arrowprops=dict(arrowstyle="->", lw=1.5),
    fontsize=11
)

plt.annotate(
    "Stable Growth",
    xy=(wages_spike_date, eci_same_date_value),
    xytext=(pd.Timestamp("2018-06-01"), eci_same_date_value - 6),
    arrowprops=dict(arrowstyle="->", lw=1.5),
    fontsize=11
)

plt.title("Composition Effect Bias: Standard Wages vs. ECI", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Index (2015 = 100)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
print('ECI:Employment Cost Index')
```
